### Import Libraries

```python
from ucimlrepo import fetch_ucirepo
```

```python
# importing modules
import tensorflow as tf
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Activation
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

### Get X (features) and Y (target) valeus from data

```python
# fetch dataset
adultdataset = fetch_ucirepo(id=763)

# data (as pandas dataframes)
X = adultdataset.data.features
y = adultdataset.data.targets
```

### Look at the data to check for missing values and datatypes

```python
X.info()
X.head()
```

```python
y.info()
y.head()
```

### Split data and create Perceptron model

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=40)

model = Sequential([Dense(2000, activation='relu'),Dense(2000, activation='relu'),Dense(2000, activation='relu'),Dense(6, activation='softmax'),])

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

model.fit(X_train, y_train, epochs=100, batch_size=13, validation_split=0.2)
```

### Evaluate the accuracy of the model

```python
acc = model.evaluate(X_test, y_test, verbose = 1)
print('Test Loss:, Test Accuracy:', acc)
```
